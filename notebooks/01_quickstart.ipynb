# Alpha 因子挖掘系统 v3.0 - 快速入门

本 Notebook 展示如何用最少的代码完成完整的因子研究工作流。

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import sys
sys.path.insert(0, '..')

# 导入 v3.0 高级 API
from alpha_mining import init_data, quick_evaluate, compare_factors
```

## 步骤 1: 一键加载数据

一行代码完成数据加载、标准化、衍生特征计算。

```python
# 加载A股数据
data = init_data(
    market='a_share',
    start='2019-01-01',
    end='2024-12-31'
)

print(f"\n数据形状: {data.shape}")
print(f"列名: {list(data.columns)}")
```

## 步骤 2: 快速评估单个因子

一行代码完成因子计算、指标评估、图表展示。

```python
# 经典均值回归因子
result = quick_evaluate(
    "rank(ts_mean(close,20)/close)",
    data,
    n_groups=5,
    neutralize=False
)
```

## 步骤 3: 批量对比多个因子

一行代码完成多个因子的批量评估和对比。

```python
# 批量对比多个因子
expressions = [
    "rank(ts_mean(close,20)/close)",  # 均值回归
    "ts_zscore(volume,10)*-1",          # 成交量异常
    "-rank(ts_std(return_1d,20))",     # 低波动异象
    "rank(close_vwap_diff)",            # 成交均价偏离
    "ts_skew(return_1d,20)*-1",        # 收益率偏度
]

results = compare_factors(expressions, data, sort_by='Rank ICIR_train')
```

## 步骤 4: 启用中性化重新评估

行业市值中性化可以提高因子的稳健性。

```python
# 启用行业市值中性化后的因子表现
result_neutral = quick_evaluate(
    "rank(ts_mean(close,20)/close)",
    data,
    neutralize=True
)
```
